```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import PolynomialFeatures

degrees = [2,3,4]

for d in degrees:


  print("R√©sultats de l'apprentissage pour degree =",d)
  polynomial_features = PolynomialFeatures(degree =  d)
  X_train_poly= polynomial_features.fit_transform(XTrainStd)
  X_test_poly = polynomial_features.fit_transform(xtestStd)
  model = LinearRegression()
  model.fit(X_train_poly,yTrain)

  rsq=sklearn.metrics.r2_score(yTrain,model.predict(X_train_poly))
  print(f"Coefficient de determination R¬≤ en prentissage (train):{rsq:.2f}")
  from sklearn.metrics import mean_squared_error

# Calcul de l'erreur quadratique moyenne sur les donn√©es d'apprentissage
  y_pred_train = model.predict(X_train_poly)
  mse_train = mean_squared_error(yTrain, y_pred_train)
  print("MSE sur les donn√©es d'apprentissage:", mse_train)

  # Calcul de l'erreur quadratique moyenne sur les donn√©es de test
  y_pred_test = model.predict(X_test_poly)
  mse_test = mean_squared_error(yTest, y_pred_test)
  print("MSE sur les donn√©es de test:", mse_test)
  print("\n")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
import tensorflow as tf
import time
import pandas as pd
import numpy as np
# Assuming 'D' is your dataset and 'Y' is your target variable
D = pd.read_table("data.txt")
data = D
Y = data["PAC"]
X = data.drop("PAC", axis=1)

XTrain, XTest, yTrain, yTest = train_test_split(X, Y, test_size=0.2, random_state=100)

# Scaling the data
scaler = MinMaxScaler()
XTrainStd = scaler.fit_transform(XTrain)
XTestStd = scaler.transform(XTest)

from keras import backend as K

# Define R¬≤ metric function
def r_squared(y_true, y_pred):
    SS_res =  K.sum(K.square(y_true - y_pred))
    SS_tot = K.sum(K.square(y_true - K.mean(y_true)))
    return (1 - SS_res/(SS_tot + K.epsilon()))

model = Sequential()
model.add(Dense(20, input_dim=15, activation='relu'))  # Adjust input_dim to match the number of features
model.add(Dense(5, input_dim=15, activation='relu'))
# Output layer for regression should not have an activation function
model.add(Dense(1))

opt = Adam(learning_rate=0.01)
model.compile(loss='mean_squared_error', optimizer=opt, metrics=['mean_squared_error','mape', r_squared])
history = model.fit(XTrainStd, yTrain, epochs=3, batch_size=6, verbose=1)  # verbose=1 to see training progress

score = model.evaluate(XTestStd, yTest)
end_time= time.time()

print("le temps d'entrainement est :",end_time-start_time)
print('Test loss:', score[0])
print('Test mean squared error:', score[1])
print('Test R¬≤:', score[2])

```

```python
import time
import numpy as np
import random
from sklearn.preprocessing import MinMaxScaler
from sklearn import model_selection

# Supposons que vous avez des listes des valeurs √† tester pour chaque param√®tre
units_values = [3,7, 11,15]
hidden_layers_values = [10, 15,20]
batch_size_values = [32, 64]
epochs_values = [3, 6,9]

# Dictionnaire pour stocker les r√©sultats
results = {}
def select_random_columns(dataframe, k):
    all_columns = list(dataframe.columns)
    selected_columns = random.sample(all_columns, k)
    return dataframe[selected_columns]

# Boucles pour tester diff√©rentes combinaisons de param√®tres
for units in units_values:
  for hidden_layers in hidden_layers_values:
        for batch_size in batch_size_values:
            for epochs in epochs_values:

                # Construction du mod√®le avec les param√®tres actuels
                data = pd.read_table("data.txt",header=0)
                Y = data["PAC"]
                data1 = data.drop("PAC", axis=1)
                X = select_random_columns(data1, units)


                XTrain,XTest,yTrain,yTest = model_selection.train_test_split(X,Y,test_size=0.2,random_state=100)


                scalar = MinMaxScaler()
                XTrainStd = scalar.fit_transform(XTrain)
                XTestStd = scalar.transform(XTest)

                # Entra√Ænement du mod√®le et mesure du temps d'entra√Ænement
                start_time = time.time()


                model = Sequential()
                model.add(Dense(hidden_layers, input_dim=units, activation='relu'))  #

                # Output layer for regression should not have an activation function
                model.add(Dense(1))

                opt = Adam(learning_rate=0.1)
                model.compile(loss='mean_squared_error', optimizer=opt, metrics=['mean_squared_error','R2Score'])
                history = model.fit(XTrainStd, yTrain, epochs=epochs, batch_size=batch_size, verbose=1)

                score = model.evaluate(XTestStd, yTest)


                end_time = time.time()
                training_time = end_time - start_time

                # Enregistrer les r√©sultats dans le dictionnaire
                results[(units, hidden_layers, batch_size, epochs)] = {
                    'R¬≤ :score_prediction': score[2] ,
                    #  score de pr√©diction obtenu,
                    'training_time': training_time}


# Affichage des r√©sultats sous forme de tableau
print("Units | Hidden Layers | Batch Size | Epochs | Score Prediction | Training Time")
for params, metrics in results.items():
    print(f"{params[0]:5} | {params[1]:13} | {params[2]:10} | {params[3]:6} | {metrics['R¬≤ :score_prediction']:15}  | {metrics['training_time']:13.4f}")

```

```python
import pandas as pd

# Chargez votre ensemble de donn√©es (remplacez 'dataset.csv' par votre nom de fichier)
data = pd.read_table('data.txt')  # Assurez-vous que votre ensemble de donn√©es est charg√© correctement

# Calcul de la matrice de corr√©lation
correlation_matrix = data.corr()

# Afficher la matrice de corr√©lation (optionnel)
print("Matrice de corr√©lation :")
print(correlation_matrix)

# S√©lection des caract√©ristiques les plus corr√©l√©es avec la puissance PV
target_column = "PAC"  # Remplacez 'Puissance PV' par le nom de votre colonne cible
threshold = 0.5  # D√©finissez un seuil de corr√©lation

relevant_features = correlation_matrix[abs(correlation_matrix[target_column]) > threshold][target_column]
relevant_features = relevant_features.drop(target_column)  # Supprimez la cible des caract√©ristiques s√©lectionn√©es

# Afficher les caract√©ristiques s√©lectionn√©es
print("\nCaract√©ristiques s√©lectionn√©es les plus corr√©l√©es avec Puissance PV :")
print(relevant_features)

# Utilisez les caract√©ristiques s√©lectionn√©es pour construire votre deuxi√®me mod√®le pr√©dictif
# ... (Utilisez les caract√©ristiques s√©lectionn√©es pour construire votre mod√®le)
```

```python
import time
import numpy as np
import random
from sklearn.preprocessing import MinMaxScaler
from sklearn import model_selection
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
import tensorflow as tf
def select_random_columns(dataframe, k):
    all_columns = list(dataframe.columns)
    selected_columns = random.sample(all_columns, k)
    return dataframe[selected_columns]


from keras import backend as K

# Define R¬≤ metric function
def r_squared(y_true, y_pred):
    SS_res =  K.sum(K.square(y_true - y_pred))
    SS_tot = K.sum(K.square(y_true - K.mean(y_true)))
    return (1 - SS_res/(SS_tot + K.epsilon()))

def mean_absolute_percentage_error(y_true, y_pred):
    epsilon = 1e-8  # A small value to avoid division by zero
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    mask = y_true != 0  # Creating a mask for non-zero true values
    return np.mean(np.abs((y_true - y_pred) / (y_true + epsilon))) * 100  # Adding epsilon to avoid division by zero

def mape(y_true, y_pred):
    return tf.py_function(mean_absolute_percentage_error, (y_true, y_pred), tf.float32)

data = pd.read_table('data.txt')
selected_features = ['Tm', 'Eff', 'RH', 'TOA', 'GHI', 'BHI', 'BNI']
X = data[selected_features]
Y = data["PAC"]

XTrain, XTest, yTrain, yTest = model_selection.train_test_split(X, Y, test_size=0.2, random_state=100)

# Scaling the data
scaler = MinMaxScaler()
XTrainStd = scaler.fit_transform(XTrain)
XTestStd = scaler.transform(XTest)

start_time = time.time()

model = Sequential()
model.add(Dense(20, input_dim=7, activation='relu'))  # Adjust input_dim to match the number of features

# Output layer for regression should not have an activation function
model.add(Dense(1))

opt = Adam(learning_rate=0.1)
model.compile(loss='mean_squared_error', optimizer=opt, metrics=['mean_squared_error', r_squared,mape])
history = model.fit(XTrainStd, yTrain, epochs=9, batch_size=32, verbose=1)  # verbose=1 to see training progress

score = model.evaluate(XTestStd, yTest)

print('Test loss:', score[0])
print('Test mean squared error:', score[1])
print('Test R¬≤:', score[2])
print('MAPE',score[3])
end_time = time.time()
training_time = end_time - start_time
print("le temps d'entra√Ænement est :",training_time)

```
